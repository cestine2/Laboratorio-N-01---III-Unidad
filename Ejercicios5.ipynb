# Seaborn example

### Introduction:

Seaborn is a library to make prety graphs in Python, see [Seaborn](https://stanford.edu/~mwaskom/software/seaborn/index.html)  


To practice, we use a dataset about tips in a restaurant

### 1. Import the necessary libraries:

```python
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
% matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

### 2. Import the dataset (tips)

```python
filecsv = "tips.csv"
tips = pd.read_csv(filecsv)

tips.head()
```

### 3. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']

tips.head()
```

### 4. Plot the total_bill column histogram

```python
# create histogram
ttbill = sns.distplot(tips['total_bill']);

# set lables and titles
ttbill.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")

# take out the right and upper borders
sns.despine()
```

### 5. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips, kind="hex")
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips, kind='kde')
```

### 6.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips, kind='reg')
```

```python
# Use different colors for gender also, diferent markers
sns.pairplot(tips, kind='reg', hue="sex")
```

### 7. Present the relationship between days and total_bill value

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True);
```

### 8. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True);
```

### 9.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x = "day", y = "tip", hue = "sex", data = tips);
```

### 10. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# better seaborn style
sns.set(style = "ticks")

# creates FacetGrid
g = sns.FacetGrid(tips, col = "sex")
g.map(plt.hist, "total_bill");
```

### 11. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)

g.add_legend();
```
